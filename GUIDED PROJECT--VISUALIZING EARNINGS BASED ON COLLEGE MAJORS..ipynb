# PANDAS AND HISTOGRAMS

Unlike in the precious exercise where we compared to columns and made use of the Dataframe as a way of selecting the columns. We can also be interested in the analysis of just one column and then we can make use of the Series.plot( ) to plot for a specific column.


Now, we will use our soon to be created histogram to answer specific questions, But first, we will generate histograms for some columns in our datset.

```python
extract = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate","Men", "Women"]
title = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate","Men", "Women"]
fig = plt.figure(figsize=(12,30))#We need to take note, here i created only the canvas for my plot, if i had
#created both the canvas and subplots, the subplots would have overlapped with the subplot creted beneath
#ax.set_xticks = ([])
#ax.set_yticks = ([])

for i in range(8):
    ax = fig.add_subplot(8,1, i+1)
    ax = recent_grads[extract[i]].hist(bins = 10, color ="green")
    ax = ax.set_xlabel(title[i])
    
```

# SCATTER MATRIX PLOTS

In the last exercises, we created individal scatter plots and histograms to analyze our data. Now, we will make use of Scatter Matrix Plot. 

A **Scatter matrix** plot combines both scatter plots and histograms into one grid of plots and allows us to explore potential relationships and distribution simultaneously. It consists of n by n plot on a grid where n is the number of columns.

Pandas contains a function named scattermatrix() that generates scatter plots. The function is part of the pandas.plotting module and needs to be imported seperately.

```python
from pandas.plotting import scatter_matrix#import scatter plot from the pandas. plotting module

#Now, we will create a scatter 2 by 2 matrix plot using Sample_size and Median columns

scatter_matrix(recent_grads[["Sample_size", "Median"]], figsize = (10,8))#Scatter matrix plots for columns: sample size
#and median
```

```python
scatter_matrix(recent_grads[["Sample_size", "Median", "Unemployment_rate"]], figsize = (10,8))#Scatter matrix plots for  
#columns: sample_size, median and unemployment
```

From the 3 by 3 scatter matrix, we can try to draw a conclusion or an observation from any pf the columns passed in. Now, lets see how umemployment rate affects Median Salary.

It is a norm that the more there are unemployed graduates in a particular major, the lesser the Median Salary collected by full time employees. Because there will be more people tussling for very few spots, hence employers of labour will tend to pay less as definitely there will be people willingly ready to pick up the jobs.

By considering the scatter plots the rightht, beginning from the second row, we are able to see the plots related to Unemplyment rate and Median. From the scatter plots, there is no correlation between unemployment rate and Median, this might be as a result of very low unemployment rate.
